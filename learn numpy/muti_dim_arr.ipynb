## multidemantional indexing

```python
multidinarr =np.array([[['A','B','C'],['D','E','F'],['G','H','I']],#index layer 0
                      [['A','B','C'],['D','E','F'],['G','H','I']],# index layer 1
                      [['A','B','C'],['D','E','F'],['G','H','I']]])#index layer 2
print(multidinarr[0,0,1])#will print at index[0,0,1]
```
