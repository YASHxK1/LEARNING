```python
import pandas as pd
```

```python
df=pd.read_csv("E:/Jupyter_notebook/learn_pandas/pokemon.csv")
print(df)
```

```python
# drop column
df2 = df.drop(columns=["Legendary"])
print(df2)
```

```python
# drop missing values 

df3 = df.dropna(subset=["Type2"])
print(df3)##al the NaN droped
```

```python
# fill missing values 
df4 = df.fillna({"Type2":"none"})
print(df4)
```

```python
# fixing values
df["Type1"]= df["Type1"].replace({"Grass":"GRASS","Fire":"FIRE"})
print(df)
```
