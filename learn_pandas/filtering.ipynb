```python
import pandas as pd
```

```python
csvdf=pd.read_csv("E:/Jupyter_notebook/learn_pandas/pokemon.csv")
print(csvdf)
```

```python
tall_pokemon=csvdf[csvdf["Height"] >= 2]
print(tall_pokemon)
```

```python
poktype = csvdf[csvdf["Legendary"] == True]
print(poktype)
```

```python
waterpok = csvdf[(csvdf["Type1"] =='Water') |
                 (csvdf["Type2"] =='Water')  ]
print(waterpok)
```

```python
pok = csvdf[(csvdf["Type1"] =='Fire') & (csvdf["Type2"] =='Flying')  ]
print(pok)
```
