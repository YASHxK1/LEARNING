## series datatypes

```python
#python list
data = [100,101,102,103]
#list to series 
series =pd.Series(data)#remember here the 'series' is constructer
print(series)
```

## seting custom lable

```python
#python list
data = [100,101,102]
#list to series 
series =pd.Series(data,index=['a','b','c']) #custom lables
print(series)
```

```python
#the loc property
print(series.loc['b'])#get value by lable/key
#-----------------------------------------
series.loc['b']=200#set value by lable/key
print(series.loc['b'])#get value by lable/key
```

```python
print(series.iloc[1]) #value by location/index
```

## filter by values

```python
data = [100,101,102,200,204]
series =pd.Series(data,index=['a','b','c','d','e']) #custom lables
```

```python
print(series[series>=200])
```

```python
print(series[series < 200])
```

```python
#newexample
calories ={"day1:": 1750,"day2:": 2100,"day3:": 1700}
series =pd.Series(calories)

print (series.loc["day1:"])
```

```python
series.loc["day3:"]=500
print (series.loc["day3:"])
```

```python
#filter by val
print(series[series >= 2000])
print(series[series >2000])
```
